```python
"""
================================================================================
NAFLD DETECTION SYSTEM - LEVEL 1 ONLY
M.Tech Project Implementation - Extended with LR & SVM
================================================================================

Task: Binary Classification - NAFLD Present (1) or Absent (0)
Algorithms: KNN, Random Forest, Decision Tree, Logistic Regression, SVM
Data Split: 4090 (Training/Validation) + 1000 (Final Testing - RESERVED)

IMPORTANT: 1000 records are RESERVED for final testing after Stage 2 is complete.
           They will NOT be used in Stage 1 evaluation.

Level 2 (Severity) will be added next week.
Final testing on 1000 records will be done after both stages are complete.

Author: [Your Name]
Date: oct 2025
================================================================================
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import joblib
import warnings
from datetime import datetime

# Machine Learning Libraries
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score, roc_curve
)

# Algorithms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# ============================================================================
# CREATE PROJECT DIRECTORIES
# ============================================================================
def create_project_structure():
    """Create necessary directories for the project"""
    directories = [
        'saved_models',
        'results',
        'plots/level1_confusion_matrices',
        'plots/level1_roc_curves',
        'plots/level1_feature_importance',
        'data'
    ]
    
    for directory in directories:
        os.makedirs(directory, exist_ok=True)
    
    print("‚úì Project structure created\n")

create_project_structure()

# ============================================================================
# HEADER
# ============================================================================
print("\n" + "="*80)
print(" NAFLD DETECTION SYSTEM - LEVEL 1 ".center(80, "="))
print("="*80)
print("\nüìä Task: NAFLD Present (1) or Absent (0)")
print("ü§ñ Algorithms: KNN, RF, DT, Logistic Regression, SVM")
print("üìÅ Data Split: 4090 (Train/Val) + 1000 (RESERVED for final testing)")
print("‚ö†Ô∏è  IMPORTANT: 1000 records will be tested ONLY after Stage 2 is complete\n")

# ============================================================================
# STEP 1: DATA LOADING
# ============================================================================
print("="*80)
print(" STEP 1: DATA LOADING ".center(80, "="))
print("="*80)

# Load dataset
df = pd.read_csv('/kaggle/input/severity-nafld/NHANES_2017_2018_full.csv')

print(f"\n‚úì Dataset loaded successfully!")
print(f"  üìÅ Total records: {len(df)}")
print(f"  üìä Total columns: {len(df.columns)}")

# Select relevant columns
selected_columns = [
    "SEQN", 
    "RIDAGEYR", 
    "RIAGENDR",
    "BMXHT", 
    "BMXWT", 
    "BMXBMI",
    "LBXSAPSI",  # ALT (U/L)
    "LBXSATSI",  # AST (U/L)
    "LBXSGTSI",  # GGT (U/L)
    "LBDTRSI",   # Triglycerides (mmol/L)
    "LBDSUASI",   # Uric Acid (¬µmol/L)   
]

# Create working dataframe
selected_df = df[selected_columns].copy()

# Rename columns for clarity
selected_df = selected_df.rename(columns={
    "SEQN": "Patient_ID",
    "RIDAGEYR": "Age",
    "RIAGENDR": "Gender",
    "BMXHT": "Height",
    "BMXWT": "Weight",
    "BMXBMI": "BMI",
    "LBXSAPSI": "ALT",
    "LBXSATSI": "AST",
    "LBXSGTSI": "GGT",
    "LBDTRSI": "Triglycerides",
    "LBDSUASI": "Uric_Acid",
})

# ============================================================================
# STEP 2: DATA PREPROCESSING
# ============================================================================
print("\n" + "="*80)
print(" STEP 2: DATA PREPROCESSING ".center(80, "="))
print("="*80)

# Check for duplicates
duplicates = selected_df.duplicated().sum()
print(f"\nüìã Duplicate records: {duplicates}")

if duplicates > 0:
    selected_df = selected_df.drop_duplicates()
    print(f"‚úì Removed {duplicates} duplicate records")

# Check missing values
print(f"\nüìä Missing values before imputation:")
missing_summary = selected_df.isnull().sum()
missing_percent = (selected_df.isnull().sum() / len(selected_df) * 100).round(2)

for col in selected_df.columns:
    if missing_summary[col] > 0:
        print(f"  ‚Ä¢ {col}: {missing_summary[col]} ({missing_percent[col]}%)")

# Fill missing values with median
numerical_cols = ["Age",'Height', 'Weight', 'BMI', 'ALT', 'AST', 
                  'GGT','Triglycerides', 'Uric_Acid']

for col in numerical_cols:
    if col in selected_df.columns:
        selected_df[col] = selected_df[col].fillna(selected_df[col].median())

print(f"\n‚úì Missing values filled with median")
print(f"‚úì Missing values after imputation: {selected_df.isnull().sum().sum()}")

# Convert Gender (1=Male‚Üí0, 2=Female‚Üí1)
selected_df['Gender'] = selected_df['Gender'].replace({1: 0, 2: 1})
print(f"\n‚úì Gender encoded: 0=Male, 1=Female")
print(f"  ‚Ä¢ Male (0): {(selected_df['Gender']==0).sum()}")
print(f"  ‚Ä¢ Female (1): {(selected_df['Gender']==1).sum()}")

# Display basic statistics
print(f"\nüìä Basic Statistics:")
print(selected_df[['Age', 'BMI', 'ALT', 'GGT', 'Triglycerides', 'Uric_Acid']].describe().round(2))

# ============================================================================
# STEP 3: NAFLD LABELING (RULE-BASED)
# ============================================================================
print("\n" + "="*80)
print(" STEP 3: NAFLD LABELING ".center(80, "="))
print("="*80)

print("\nüìã Clinical Rules for NAFLD Detection:")
print("   Risk factors (need ‚â•3 out of 5):")
print("   1. BMI ‚â• 25")
print("   2. ALT > 30 (Male) or > 19 (Female)")
print("   3. GGT > 26")
print("   4. Triglycerides ‚â• 1.7 mmol/L")
print("   5. Uric Acid ‚â• 320 ¬µmol/L (Male) or ‚â• 288 ¬µmol/L (Female)")

def nafld_present(row):
    """
    Determine if NAFLD is present based on clinical criteria
    Returns: 1 if NAFLD present, 0 if absent
    """
    count = 0
    
    # Risk Factor 1: BMI
    if row['BMI'] >= 25:
        count += 1
    
    # Risk Factor 2: ALT (gender-specific)
    if (row['Gender'] == 0 and row['ALT'] > 30) or (row['Gender'] == 1 and row['ALT'] > 19):
        count += 1
    
    # Risk Factor 3: GGT
    if row['GGT'] > 26:
        count += 1
    
    # Risk Factor 4: Triglycerides
    if row['Triglycerides'] >= 1.7:
        count += 1
    
    # Risk Factor 5: Uric Acid (gender-specific)
    if (row['Gender'] == 0 and row['Uric_Acid'] >= 320) or (row['Gender'] == 1 and row['Uric_Acid'] >= 288):
        count += 1
    
    # NAFLD Present if at least 3 risk factors
    return 1 if count >= 3 else 0

# Apply labeling
selected_df['NAFLD_Status'] = selected_df.apply(nafld_present, axis=1)

# Check distribution
nafld_counts = selected_df['NAFLD_Status'].value_counts()
nafld_percent = selected_df['NAFLD_Status'].value_counts(normalize=True) * 100

print(f"\n‚úì NAFLD labeling completed")
print(f"\nüìä NAFLD Status Distribution:")
print(f"  ‚Ä¢ Absent (0): {nafld_counts[0]} ({nafld_percent[0]:.2f}%)")
print(f"  ‚Ä¢ Present (1): {nafld_counts[1]} ({nafld_percent[1]:.2f}%)")

# ============================================================================
# STEP 4: DATA SPLITTING
# ============================================================================
print("\n" + "="*80)
print(" STEP 4: DATA SPLITTING ".center(80, "="))
print("="*80)

print(f"\nüîÄ Splitting Strategy:")
print(f"  Total: {len(selected_df)} records")
print(f"  ‚Ä¢ 4090 ‚Üí Training & Validation (for Stage 1 & 2)")
print(f"  ‚Ä¢ 1000 ‚Üí RESERVED for final testing (after both stages complete)")

# First split: 4090 for training, 1000 for final testing (RESERVED)
train_val_data, final_test_data = train_test_split(
    selected_df, 
    test_size=1000, 
    random_state=42,
    stratify=selected_df['NAFLD_Status']
)

print(f"\n‚úì First split:")
print(f"  ‚Ä¢ Train/Val: {len(train_val_data)} records")
print(f"  ‚Ä¢ Final Test (RESERVED): {len(final_test_data)} records")

# Second split: 80% train, 20% validation from 4090
train_data, val_data = train_test_split(
    train_val_data,
    test_size=0.2,
    random_state=42,
    stratify=train_val_data['NAFLD_Status']
)

print(f"\n‚úì Second split (from 4090):")
print(f"  ‚Ä¢ Training: {len(train_data)} records")
print(f"  ‚Ä¢ Validation: {len(val_data)} records")

# Check class distribution in splits
print(f"\nüìä Class Distribution Check:")
print(f"  Training - Absent: {(train_data['NAFLD_Status']==0).sum()}, Present: {(train_data['NAFLD_Status']==1).sum()}")
print(f"  Validation - Absent: {(val_data['NAFLD_Status']==0).sum()}, Present: {(val_data['NAFLD_Status']==1).sum()}")
print(f"  Final Test - Absent: {(final_test_data['NAFLD_Status']==0).sum()}, Present: {(final_test_data['NAFLD_Status']==1).sum()}")

# Save datasets
train_data.to_csv('data/train_data.csv', index=False)
val_data.to_csv('data/val_data.csv', index=False)
final_test_data.to_csv('data/final_test_1000_RESERVED.csv', index=False)
print(f"\n‚úì Datasets saved to 'data/' directory")
print(f"  ‚ö†Ô∏è  'final_test_1000_RESERVED.csv' will be used ONLY after Stage 2")

# ============================================================================
# STEP 5: FEATURE PREPARATION
# ============================================================================
print("\n" + "="*80)
print(" STEP 5: FEATURE PREPARATION ".center(80, "="))
print("="*80)

# Selected features (7 features)
features = ['Gender', 'Age', 'BMI', 'ALT', 'GGT', 'Triglycerides', 'Uric_Acid']
target = 'NAFLD_Status'

print(f"\nüéØ Selected Features (7 features):")
for i, feature in enumerate(features, 1):
    print(f"  {i}. {feature}")

# Prepare datasets (ONLY train and validation - NO final test)
X_train = train_data[features]
y_train = train_data[target]

X_val = val_data[features]
y_val = val_data[target]

print(f"\nüìä Data Shapes:")
print(f"  ‚Ä¢ X_train: {X_train.shape}")
print(f"  ‚Ä¢ X_val: {X_val.shape}")
print(f"  ‚Ä¢ Final test data: RESERVED (not loaded)")

# ============================================================================
# STEP 6: FEATURE SCALING
# ============================================================================
print("\n" + "="*80)
print(" STEP 6: FEATURE SCALING ".center(80, "="))
print("="*80)

print("\n‚öôÔ∏è Applying StandardScaler (important for KNN, LR, and SVM)...")

# Initialize and fit scaler on training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

print(f"‚úì Features scaled")
print(f"  ‚Ä¢ Mean: ~0, Std: ~1")

# Save scaler for future use
joblib.dump(scaler, 'saved_models/scaler_level1.pkl')
print(f"‚úì Scaler saved: 'saved_models/scaler_level1.pkl'")

# ============================================================================
# STEP 7: MODEL TRAINING
# ============================================================================
print("\n" + "="*80)
print(" STEP 7: TRAINING MODELS ".center(80, "="))
print("="*80)

import time

def train_model_with_timer(model, X_train, y_train, model_name):
    """Train model and track time"""
    print(f"\nüîÑ Training {model_name}...")
    start_time = time.time()
    model.fit(X_train, y_train)
    elapsed = time.time() - start_time
    print(f"   ‚úì Completed in {elapsed:.2f} seconds")
    return model

# Initialize models
print("\nü§ñ Initializing models...")

# 1. K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5, weights='uniform', metric='euclidean')

# 2. Random Forest
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=20,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

# 3. Decision Tree
dt_model = DecisionTreeClassifier(
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    criterion='gini',
    random_state=42
)

# 4. Logistic Regression
lr_model = LogisticRegression(
    max_iter=1000,
    random_state=42,
    solver='lbfgs',
    C=1.0
)

# 5. Support Vector Machine (SVM)
svm_model = SVC(
    kernel='rbf',
    C=1.0,
    gamma='scale',
    random_state=42,
    probability=True  # Enable probability estimates for ROC curve
)

# Train all models
knn_model = train_model_with_timer(knn_model, X_train_scaled, y_train, "KNN")
rf_model = train_model_with_timer(rf_model, X_train_scaled, y_train, "Random Forest")
dt_model = train_model_with_timer(dt_model, X_train_scaled, y_train, "Decision Tree")
lr_model = train_model_with_timer(lr_model, X_train_scaled, y_train, "Logistic Regression")
svm_model = train_model_with_timer(svm_model, X_train_scaled, y_train, "SVM")

print(f"\n‚úì All models trained successfully!")

# Save models
joblib.dump(knn_model, 'saved_models/level1_knn.pkl')
joblib.dump(rf_model, 'saved_models/level1_rf.pkl')
joblib.dump(dt_model, 'saved_models/level1_dt.pkl')
joblib.dump(lr_model, 'saved_models/level1_lr.pkl')
joblib.dump(svm_model, 'saved_models/level1_svm.pkl')
print(f"‚úì Models saved in 'saved_models/' directory")

# ============================================================================
# STEP 8: MODEL EVALUATION ON VALIDATION SET ONLY
# ============================================================================
print("\n" + "="*80)
print(" STEP 8: MODEL EVALUATION (Validation Set ONLY) ".center(80, "="))
print("="*80)

print("\n‚ö†Ô∏è  IMPORTANT: Evaluating on VALIDATION SET only")
print("   Final test (1000 records) will be used after Stage 2 is complete\n")

def evaluate_model_detailed(model, X, y, model_name, dataset_name="Validation"):
    """Comprehensive evaluation function"""
    
    # Predictions
    y_pred = model.predict(X)
    y_pred_proba = model.predict_proba(X)[:, 1] if hasattr(model, 'predict_proba') else None
    
    # Metrics
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred, average='binary', zero_division=0)
    recall = recall_score(y, y_pred, average='binary', zero_division=0)
    f1 = f1_score(y, y_pred, average='binary', zero_division=0)
    
    print(f"\n{'='*60}")
    print(f" {model_name} - {dataset_name} Set ".center(60))
    print(f"{'='*60}")
    print(f"  Accuracy:  {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1-Score:  {f1:.4f}")
    
    # ROC-AUC if probability available
    if y_pred_proba is not None:
        roc_auc = roc_auc_score(y, y_pred_proba)
        print(f"  ROC-AUC:   {roc_auc:.4f}")
    
    # Classification Report
    print(f"\n  Classification Report:")
    print(classification_report(y, y_pred, target_names=['Absent (0)', 'Present (1)'], digits=4))
    
    # Confusion Matrix
    cm = confusion_matrix(y, y_pred)
    
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Absent', 'Present'],
                yticklabels=['Absent', 'Present'])
    plt.title(f'{model_name} - Confusion Matrix\n{dataset_name} Set')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.tight_layout()
    plt.savefig(f'plots/level1_confusion_matrices/{model_name.lower().replace(" ", "_")}_{dataset_name.lower()}.png', dpi=300)
    plt.show()
    
    return {
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'ROC-AUC': roc_auc if y_pred_proba is not None else None
    }

# Evaluate all models on validation set ONLY
results_val = []
results_val.append(evaluate_model_detailed(knn_model, X_val_scaled, y_val, "KNN", "Validation"))
results_val.append(evaluate_model_detailed(rf_model, X_val_scaled, y_val, "Random Forest", "Validation"))
results_val.append(evaluate_model_detailed(dt_model, X_val_scaled, y_val, "Decision Tree", "Validation"))
results_val.append(evaluate_model_detailed(lr_model, X_val_scaled, y_val, "Logistic Regression", "Validation"))
results_val.append(evaluate_model_detailed(svm_model, X_val_scaled, y_val, "SVM", "Validation"))

# ============================================================================
# STEP 9: MODEL COMPARISON
# ============================================================================
print("\n" + "="*80)
print(" STEP 9: MODEL COMPARISON ".center(80, "="))
print("="*80)

# Create comparison DataFrame
comparison_df = pd.DataFrame(results_val)
comparison_df = comparison_df.set_index('Model')

print(f"\nüìä Model Performance Comparison (Validation Set):")
print(comparison_df.round(4))

# Find best model
best_model_name = comparison_df['Accuracy'].idxmax()
best_accuracy = comparison_df['Accuracy'].max()

print(f"\nüèÜ Best Model: {best_model_name}")
print(f"   Accuracy: {best_accuracy:.4f} ({best_accuracy*100:.2f}%)")

# Save comparison
comparison_df.to_csv('results/level1_model_comparison_validation.csv')
print(f"\n‚úì Comparison saved: 'results/level1_model_comparison_validation.csv'")

# Visualize comparison
fig, axes = plt.subplots(1, 4, figsize=(18, 4))

metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
colors_palette = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#f39c12']

for idx, metric in enumerate(metrics):
    axes[idx].bar(comparison_df.index, comparison_df[metric], color=colors_palette)
    axes[idx].set_title(f'{metric}', fontweight='bold')
    axes[idx].set_ylim([0.90, 1.03])
    axes[idx].set_ylabel('Score')
    axes[idx].tick_params(axis='x', rotation=45)
    
    # Add value labels on bars
    for i, v in enumerate(comparison_df[metric]):
        axes[idx].text(i, v + 0.01, f'{v:.3f}', ha='center', fontsize=8)

plt.tight_layout()
plt.savefig('plots/level1_model_comparison_validation.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# STEP 10: FEATURE IMPORTANCE (for RF, DT, and LR)
# ============================================================================
print("\n" + "="*80)
print(" STEP 10: FEATURE IMPORTANCE ".center(80, "="))
print("="*80)

def plot_feature_importance(model, feature_names, model_name):
    """Plot feature importance for tree-based models"""
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]
        
        print(f"\nüìä Feature Importance - {model_name}:")
        for i, idx in enumerate(indices, 1):
            print(f"  {i}. {feature_names[idx]}: {importances[idx]:.4f}")
        
        plt.figure(figsize=(10, 6))
        plt.bar(range(len(importances)), importances[indices], color='steelblue')
        plt.xticks(range(len(importances)), [feature_names[i] for i in indices], rotation=45, ha='right')
        plt.xlabel('Features')
        plt.ylabel('Importance')
        plt.title(f'Feature Importance - {model_name}')
        plt.tight_layout()
        plt.savefig(f'plots/level1_feature_importance/{model_name.lower().replace(" ", "_")}_importance.png', dpi=300)
        plt.show()

def plot_lr_coefficients(model, feature_names, model_name):
    """Plot coefficients for Logistic Regression"""
    coefficients = model.coef_[0]
    indices = np.argsort(np.abs(coefficients))[::-1]
    
    print(f"\nüìä Feature Coefficients - {model_name}:")
    for i, idx in enumerate(indices, 1):
        print(f"  {i}. {feature_names[idx]}: {coefficients[idx]:.4f}")
    
    plt.figure(figsize=(10, 6))
    colors = ['green' if c > 0 else 'red' for c in coefficients[indices]]
    plt.bar(range(len(coefficients)), coefficients[indices], color=colors)
    plt.xticks(range(len(coefficients)), [feature_names[i] for i in indices], rotation=45, ha='right')
    plt.xlabel('Features')
    plt.ylabel('Coefficient Value')
    plt.title(f'Feature Coefficients - {model_name}')
    plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    plt.tight_layout()
    plt.savefig(f'plots/level1_feature_importance/{model_name.lower().replace(" ", "_")}_coefficients.png', dpi=300)
    plt.show()

# Plot for Random Forest and Decision Tree
plot_feature_importance(rf_model, features, "Random Forest")
plot_feature_importance(dt_model, features, "Decision Tree")

# Plot for Logistic Regression
plot_lr_coefficients(lr_model, features, "Logistic Regression")

# ============================================================================
# STEP 11: ROC CURVES (Validation Set)
# ============================================================================
print("\n" + "="*80)
print(" STEP 11: ROC CURVES (Validation Set) ".center(80, "="))
print("="*80)

plt.figure(figsize=(10, 8))

models = {
    'KNN': knn_model,
    'Random Forest': rf_model,
    'Decision Tree': dt_model,
    'Logistic Regression': lr_model,
    'SVM': svm_model
}

colors = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#f39c12']

for (model_name, model), color in zip(models.items(), colors):
    y_pred_proba = model.predict_proba(X_val_scaled)[:, 1]
    fpr, tpr, _ = roc_curve(y_val, y_pred_proba)
    roc_auc = roc_auc_score(y_val, y_pred_proba)
    
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.3f})', 
             color=color, linewidth=2)

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier', linewidth=1)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curves - All Models (Validation Set)', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('plots/level1_roc_curves/roc_curves_validation.png', dpi=300)
plt.show()

print(f"‚úì ROC curves saved")

# ============================================================================
# STEP 12: SAVE BEST MODEL INFO
# ============================================================================
print("\n" + "="*80)
print(" STEP 12: SAVE BEST MODEL INFO ".center(80, "="))
print("="*80)

# Save best model name
best_model_info = {
    'best_model_name': best_model_name,
    'best_accuracy': best_accuracy,
    'all_results': comparison_df.to_dict()
}

import json
with open('results/level1_best_model_info.json', 'w') as f:
    json.dump(best_model_info, f, indent=4)

print(f"\n‚úì Best model info saved: 'results/level1_best_model_info.json'")
print(f"  Best Model: {best_model_name}")
print(f"  Validation Accuracy: {best_accuracy:.4f}")
# ============================================================================
# STEP 12: SHAP ANALYSIS FOR EXPLAINABILITY
# ============================================================================
print("\n" + "="*80)
print(" STEP 12: SHAP ANALYSIS FOR EXPLAINABILITY ".center(80, "="))
print("="*80)

# Install SHAP if not already installed
# !pip install shap

import shap

# Create directory for SHAP plots
os.makedirs('plots/level1_shap_analysis', exist_ok=True)

print("\nüîç Generating SHAP explanations for model interpretability...")

# ============================================================================
# SHAP for Random Forest (Best performing tree-based model)
# ============================================================================
print("\n" + "-"*60)
print(" SHAP Analysis: Random Forest ".center(60))
print("-"*60)

# Create SHAP explainer for Random Forest
print("\n‚è≥ Creating SHAP TreeExplainer for Random Forest...")
explainer_rf = shap.TreeExplainer(rf_model)

# Calculate SHAP values for validation set
print("‚è≥ Calculating SHAP values (this may take a moment)...")
shap_values_rf = explainer_rf.shap_values(X_val_scaled)

# For binary classification, shap_values might be a list
# We want SHAP values for class 1 (NAFLD Present)
if isinstance(shap_values_rf, list):
    shap_values_rf_class1 = shap_values_rf[1]
else:
    shap_values_rf_class1 = shap_values_rf

print("‚úì SHAP values calculated")

# 1. Summary Plot (Bar) - Feature Importance
print("\nüìä Generating SHAP Summary Plot (Feature Importance)...")
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values_rf_class1, X_val_scaled, 
                  feature_names=features, plot_type="bar", show=False)
plt.title('SHAP Feature Importance - Random Forest\n(Stage 1: NAFLD Detection)', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('plots/level1_shap_analysis/shap_feature_importance_rf.png', 
            dpi=300, bbox_inches='tight')
plt.show()

# 2. Summary Plot (Beeswarm) - Feature Impact
print("üìä Generating SHAP Beeswarm Plot...")
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values_rf_class1, X_val_scaled, 
                  feature_names=features, show=False)
plt.title('SHAP Summary Plot - Random Forest\n(Stage 1: NAFLD Detection)', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('plots/level1_shap_analysis/shap_summary_beeswarm_rf.png', 
            dpi=300, bbox_inches='tight')
plt.show()

# 3. Mean Absolute SHAP Values
print("\nüìä Top Features by Mean Absolute SHAP Value:")
mean_abs_shap = np.abs(shap_values_rf_class1).mean(axis=0)
feature_importance_df = pd.DataFrame({
    'Feature': features,
    'Mean |SHAP|': mean_abs_shap
}).sort_values('Mean |SHAP|', ascending=False)

print(feature_importance_df.to_string(index=False))

# Save to CSV
feature_importance_df.to_csv('results/shap_feature_importance_rf.csv', index=False)
print("\n‚úì SHAP feature importance saved: 'results/shap_feature_importance_rf.csv'")

# 4. Top 3 Most Important Features
print("\nüéØ Top 3 Most Influential Features:")
top_3_features = feature_importance_df.head(3)
for idx, row in top_3_features.iterrows():
    print(f"  {idx+1}. {row['Feature']}: {row['Mean |SHAP|']:.4f}")

# ============================================================================
# SHAP for Decision Tree
# ============================================================================
print("\n" + "-"*60)
print(" SHAP Analysis: Decision Tree ".center(60))
print("-"*60)

print("\n‚è≥ Creating SHAP TreeExplainer for Decision Tree...")
explainer_dt = shap.TreeExplainer(dt_model)
shap_values_dt = explainer_dt.shap_values(X_val_scaled)

if isinstance(shap_values_dt, list):
    shap_values_dt_class1 = shap_values_dt[1]
else:
    shap_values_dt_class1 = shap_values_dt

print("‚úì SHAP values calculated")

# Summary Plot for Decision Tree
print("üìä Generating SHAP Feature Importance Plot for Decision Tree...")
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values_dt_class1, X_val_scaled, 
                  feature_names=features, plot_type="bar", show=False)
plt.title('SHAP Feature Importance - Decision Tree\n(Stage 1: NAFLD Detection)', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('plots/level1_shap_analysis/shap_feature_importance_dt.png', 
            dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# SHAP for Logistic Regression (using KernelExplainer)
# ============================================================================
print("\n" + "-"*60)
print(" SHAP Analysis: Logistic Regression ".center(60))
print("-"*60)

print("\n‚è≥ Creating SHAP LinearExplainer for Logistic Regression...")
# For linear models, use LinearExplainer (faster than KernelExplainer)
explainer_lr = shap.LinearExplainer(lr_model, X_val_scaled, feature_perturbation="interventional")
shap_values_lr = explainer_lr.shap_values(X_val_scaled)

print("‚úì SHAP values calculated")

# Summary Plot for Logistic Regression
print("üìä Generating SHAP Feature Importance Plot for Logistic Regression...")
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values_lr, X_val_scaled, 
                  feature_names=features, plot_type="bar", show=False)
plt.title('SHAP Feature Importance - Logistic Regression\n(Stage 1: NAFLD Detection)', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('plots/level1_shap_analysis/shap_feature_importance_lr.png', 
            dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# Combined SHAP Comparison
# ============================================================================
print("\n" + "-"*60)
print(" SHAP Feature Importance Comparison ".center(60))
print("-"*60)

# Calculate mean absolute SHAP for all models
mean_shap_rf = np.abs(shap_values_rf_class1).mean(axis=0)
mean_shap_dt = np.abs(shap_values_dt_class1).mean(axis=0)
mean_shap_lr = np.abs(shap_values_lr).mean(axis=0)

# Create comparison DataFrame
shap_comparison = pd.DataFrame({
    'Feature': features,
    'Random Forest': mean_shap_rf,
    'Decision Tree': mean_shap_dt,
    'Logistic Regression': mean_shap_lr
})

# Sort by Random Forest importance
shap_comparison = shap_comparison.sort_values('Random Forest', ascending=False)

print("\nüìä SHAP Feature Importance Comparison (Mean |SHAP|):")
print(shap_comparison.to_string(index=False))

# Save comparison
shap_comparison.to_csv('results/shap_comparison_all_models.csv', index=False)
print("\n‚úì SHAP comparison saved: 'results/shap_comparison_all_models.csv'")

# Visualize comparison
fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(len(features))
width = 0.25

bars1 = ax.bar(x - width, shap_comparison['Random Forest'], width, 
               label='Random Forest', color='#2ecc71')
bars2 = ax.bar(x, shap_comparison['Decision Tree'], width, 
               label='Decision Tree', color='#e74c3c')
bars3 = ax.bar(x + width, shap_comparison['Logistic Regression'], width, 
               label='Logistic Regression', color='#3498db')

ax.set_xlabel('Features', fontsize=12, fontweight='bold')
ax.set_ylabel('Mean |SHAP Value|', fontsize=12, fontweight='bold')
ax.set_title('SHAP Feature Importance Comparison Across Models\n(Stage 1: NAFLD Detection)', 
             fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(shap_comparison['Feature'], rotation=45, ha='right')
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('plots/level1_shap_analysis/shap_comparison_all_models.png', 
            dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# Individual Prediction Explanation (Sample)
# ============================================================================
print("\n" + "-"*60)
print(" Sample Individual Prediction Explanation ".center(60))
print("-"*60)

# Select a few sample predictions to explain
sample_indices = [0, 50, 100]  # First, middle, and later samples

print("\nüìä Generating waterfall plots for sample predictions...")

for idx in sample_indices:
    sample = X_val_scaled[idx:idx+1]
    sample_shap = shap_values_rf_class1[idx]
    
    # Get actual prediction
    pred = rf_model.predict(sample)[0]
    pred_proba = rf_model.predict_proba(sample)[0]
    actual = y_val.iloc[idx]
    
    print(f"\nSample {idx}: Actual={actual}, Predicted={pred}, Prob(NAFLD)={pred_proba[1]:.3f}")
    
    # Waterfall plot
    plt.figure(figsize=(10, 6))
    shap.waterfall_plot(shap.Explanation(values=sample_shap, 
                                         base_values=explainer_rf.expected_value[1] if isinstance(explainer_rf.expected_value, list) else explainer_rf.expected_value,
                                         data=sample[0], 
                                         feature_names=features), 
                       show=False)
    plt.title(f'SHAP Waterfall Plot - Sample {idx}\nActual: {actual}, Predicted: {pred}', 
              fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'plots/level1_shap_analysis/shap_waterfall_sample_{idx}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

print("\n‚úì Sample explanations generated")

# ============================================================================
# SHAP Summary
# ============================================================================
print("\n" + "="*80)
print(" SHAP ANALYSIS COMPLETE ".center(80, "="))
print("="*80)

print(f"\n‚úÖ SHAP Analysis Complete!")
print(f"\nüìä Generated Plots:")
print(f"  ‚Ä¢ SHAP feature importance (RF, DT, LR)")
print(f"  ‚Ä¢ SHAP beeswarm plot (RF)")
print(f"  ‚Ä¢ SHAP comparison across models")
print(f"  ‚Ä¢ Individual prediction explanations (samples)")
print(f"\nüìÅ Saved Files:")
print(f"  ‚Ä¢ Plots: plots/level1_shap_analysis/*.png")
print(f"  ‚Ä¢ Results: results/shap_feature_importance_rf.csv")
print(f"  ‚Ä¢ Comparison: results/shap_comparison_all_models.csv")
print(f"\nüéØ Key Insights:")
print(f"  Top 3 Most Important Features (Random Forest):")
for idx, row in top_3_features.iterrows():
    print(f"    {idx+1}. {row['Feature']}: {row['Mean |SHAP|']:.4f}")

print("\n" + "="*80)
# ============================================================================
# FINAL SUMMARY
# ============================================================================
print("\n" + "="*80)
print(" STAGE 1 COMPLETE ".center(80, "="))
print("="*80)

print(f"\n‚úÖ Stage 1 Training Complete!")
print(f"\nüìä Summary:")
print(f"  ‚Ä¢ Training records used: {len(train_data)}")
print(f"  ‚Ä¢ Validation records used: {len(val_data)}")
print(f"  ‚Ä¢ Models trained: KNN, RF, DT, Logistic Regression, SVM")
print(f"  ‚Ä¢ Best model: {best_model_name} ({best_accuracy*100:.2f}%)")
print(f"\n‚ö†Ô∏è  RESERVED for Later:")
print(f"  ‚Ä¢ Final test set: 1000 records (NOT USED YET)")
print(f"  ‚Ä¢ Will be tested AFTER Stage 2 is complete")
print(f"\nüìÅ Saved Files:")
print(f"  ‚Ä¢ Models: saved_models/level1_*.pkl (5 models)")
print(f"  ‚Ä¢ Scaler: saved_models/scaler_level1.pkl")
print(f"  ‚Ä¢ Results: results/level1_model_comparison_validation.csv")
print(f"  ‚Ä¢ Plots: plots/level1_*/")
print(f"  ‚Ä¢ Reserved test data: data/final_test_1000_RESERVED.csv")
print(f"\n‚è≠Ô∏è  Next Steps:")
print(f"  1. Build Stage 2 (Severity Classification)")
print(f"  2. Integrate Stage 1 + Stage 2 into hierarchical system")
print(f"  3. Test complete system on 1000 reserved records")
print(f"  4. Add explainability (SHAP)")

print("\n" + "="*80)
print(" END OF STAGE 1 ".center(80, "="))
print("="*80)
```

```python
"""
================================================================================
NAFLD DETECTION SYSTEM - LEVEL 1 ONLY
M.Tech Project Implementation - Corrected Version
================================================================================

Task: Binary Classification - NAFLD Present (1) or Absent (0)
Algorithms: KNN, Random Forest, Decision Tree
Data Split: 4090 (Training/Validation) + 1000 (Final Testing - RESERVED)

IMPORTANT: 1000 records are RESERVED for final testing after Stage 2 is complete.
           They will NOT be used in Stage 1 evaluation.

Level 2 (Severity) will be added next week.
Final testing on 1000 records will be done after both stages are complete.

Author: [Your Name]
Date: oct 2025
================================================================================
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import joblib
import warnings
from datetime import datetime

# Machine Learning Libraries
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score, roc_curve
)

# Algorithms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# ============================================================================
# CREATE PROJECT DIRECTORIES
# ============================================================================
def create_project_structure():
    """Create necessary directories for the project"""
    directories = [
        'saved_models',
        'results',
        'plots/level1_confusion_matrices',
        'plots/level1_roc_curves',
        'plots/level1_feature_importance',
        'data'
    ]
    
    for directory in directories:
        os.makedirs(directory, exist_ok=True)
    
    print("‚úì Project structure created\n")

create_project_structure()

# ============================================================================
# HEADER
# ============================================================================
print("\n" + "="*80)
print(" NAFLD DETECTION SYSTEM - LEVEL 1 ".center(80, "="))
print("="*80)
print("\nüìä Task: NAFLD Present (1) or Absent (0)")
print("ü§ñ Algorithms: KNN, Random Forest, Decision Tree")
print("üìÅ Data Split: 4090 (Train/Val) + 1000 (RESERVED for final testing)")
print("‚ö†Ô∏è  IMPORTANT: 1000 records will be tested ONLY after Stage 2 is complete\n")

# ============================================================================
# STEP 1: DATA LOADING
# ============================================================================
print("="*80)
print(" STEP 1: DATA LOADING ".center(80, "="))
print("="*80)

# Load dataset
df = pd.read_csv('/kaggle/input/severity-nafld/NHANES_2017_2018_full.csv')

print(f"\n‚úì Dataset loaded successfully!")
print(f"  üìÅ Total records: {len(df)}")
print(f"  üìä Total columns: {len(df.columns)}")

# Select relevant columns
selected_columns = [
    "SEQN", 
    "RIDAGEYR", 
    "RIAGENDR",
    "BMXHT", 
    "BMXWT", 
    "BMXBMI",
    "LBXSAPSI",  # ALT (U/L)
    "LBXSATSI",  # AST (U/L)
    "LBXSGTSI",  # GGT (U/L)
    "LBDTRSI",   # Triglycerides (mmol/L)
    "LBDSUASI",   # Uric Acid (¬µmol/L)   
]

# Create working dataframe
selected_df = df[selected_columns].copy()

# Rename columns for clarity
selected_df = selected_df.rename(columns={
    "SEQN": "Patient_ID",
    "RIDAGEYR": "Age",
    "RIAGENDR": "Gender",
    "BMXHT": "Height",
    "BMXWT": "Weight",
    "BMXBMI": "BMI",
    "LBXSAPSI": "ALT",
    "LBXSATSI": "AST",
    "LBXSGTSI": "GGT",
    "LBDTRSI": "Triglycerides",
    "LBDSUASI": "Uric_Acid",
})

import matplotlib.pyplot as plt
import seaborn as sns

# # Get number of unique values
# unique_counts = features.nunique()

# # Plot
# plt.figure(figsize=(12,6))
# sns.barplot(x=unique_counts.index, y=unique_counts.values, palette='viridis')
# plt.title('Number of Unique Values per Feature')
# plt.xlabel('Feature')
# plt.ylabel('Unique Values')
# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()

# print(f"‚úì Columns renamed")

# ============================================================================
# STEP 2: DATA PREPROCESSING
# ============================================================================
print("\n" + "="*80)
print(" STEP 2: DATA PREPROCESSING ".center(80, "="))
print("="*80)

# Check for duplicates
duplicates = selected_df.duplicated().sum()
print(f"\nüìã Duplicate records: {duplicates}")

if duplicates > 0:
    selected_df = selected_df.drop_duplicates()
    print(f"‚úì Removed {duplicates} duplicate records")

# Check missing values
print(f"\nüìä Missing values before imputation:")
missing_summary = selected_df.isnull().sum()
missing_percent = (selected_df.isnull().sum() / len(selected_df) * 100).round(2)

for col in selected_df.columns:
    if missing_summary[col] > 0:
        print(f"  ‚Ä¢ {col}: {missing_summary[col]} ({missing_percent[col]}%)")

# Fill missing values with median
numerical_cols = ["Age",'Height', 'Weight', 'BMI', 'ALT', 'AST', 
                  'GGT','Triglycerides', 'Uric_Acid']

for col in numerical_cols:
    if col in selected_df.columns:
        selected_df[col] = selected_df[col].fillna(selected_df[col].median())

print(f"\n‚úì Missing values filled with median")
print(f"‚úì Missing values after imputation: {selected_df.isnull().sum().sum()}")

# Convert Gender (1=Male‚Üí0, 2=Female‚Üí1)
selected_df['Gender'] = selected_df['Gender'].replace({1: 0, 2: 1})
print(f"\n‚úì Gender encoded: 0=Male, 1=Female")
print(f"  ‚Ä¢ Male (0): {(selected_df['Gender']==0).sum()}")
print(f"  ‚Ä¢ Female (1): {(selected_df['Gender']==1).sum()}")

# Display basic statistics
print(f"\nüìä Basic Statistics:")
print(selected_df[['Age', 'BMI', 'ALT', 'GGT', 'Triglycerides', 'Uric_Acid']].describe().round(2))

# ============================================================================
# STEP 3: NAFLD LABELING (RULE-BASED)
# ============================================================================
print("\n" + "="*80)
print(" STEP 3: NAFLD LABELING ".center(80, "="))
print("="*80)

print("\nüìã Clinical Rules for NAFLD Detection:")
print("   Risk factors (need ‚â•3 out of 5):")
print("   1. BMI ‚â• 25")
print("   2. ALT > 30 (Male) or > 19 (Female)")
print("   3. GGT > 26")
print("   4. Triglycerides ‚â• 1.7 mmol/L")
print("   5. Uric Acid ‚â• 320 ¬µmol/L (Male) or ‚â• 288 ¬µmol/L (Female)")

def nafld_present(row):
    """
    Determine if NAFLD is present based on clinical criteria
    Returns: 1 if NAFLD present, 0 if absent
    """
    count = 0
    
    # Risk Factor 1: BMI
    if row['BMI'] >= 25:
        count += 1
    
    # Risk Factor 2: ALT (gender-specific)
    if (row['Gender'] == 0 and row['ALT'] > 30) or (row['Gender'] == 1 and row['ALT'] > 19):
        count += 1
    
    # Risk Factor 3: GGT
    if row['GGT'] > 26:
        count += 1
    
    # Risk Factor 4: Triglycerides
    if row['Triglycerides'] >= 1.7:
        count += 1
    
    # Risk Factor 5: Uric Acid (gender-specific)
    if (row['Gender'] == 0 and row['Uric_Acid'] >= 320) or (row['Gender'] == 1 and row['Uric_Acid'] >= 288):
        count += 1
    
    # NAFLD Present if at least 3 risk factors
    return 1 if count >= 3 else 0

# Apply labeling
selected_df['NAFLD_Status'] = selected_df.apply(nafld_present, axis=1)

# Check distribution
nafld_counts = selected_df['NAFLD_Status'].value_counts()
nafld_percent = selected_df['NAFLD_Status'].value_counts(normalize=True) * 100

print(f"\n‚úì NAFLD labeling completed")
print(f"\nüìä NAFLD Status Distribution:")
print(f"  ‚Ä¢ Absent (0): {nafld_counts[0]} ({nafld_percent[0]:.2f}%)")
print(f"  ‚Ä¢ Present (1): {nafld_counts[1]} ({nafld_percent[1]:.2f}%)")

# ============================================================================
# STEP 4: DATA SPLITTING
# ============================================================================
print("\n" + "="*80)
print(" STEP 4: DATA SPLITTING ".center(80, "="))
print("="*80)

print(f"\nüîÄ Splitting Strategy:")
print(f"  Total: {len(selected_df)} records")
print(f"  ‚Ä¢ 4090 ‚Üí Training & Validation (for Stage 1 & 2)")
print(f"  ‚Ä¢ 1000 ‚Üí RESERVED for final testing (after both stages complete)")

# First split: 4090 for training, 1000 for final testing (RESERVED)
train_val_data, final_test_data = train_test_split(
    selected_df, 
    test_size=1000, 
    random_state=42,
    stratify=selected_df['NAFLD_Status']
)

print(f"\n‚úì First split:")
print(f"  ‚Ä¢ Train/Val: {len(train_val_data)} records")
print(f"  ‚Ä¢ Final Test (RESERVED): {len(final_test_data)} records")

# Second split: 80% train, 20% validation from 4090
train_data, val_data = train_test_split(
    train_val_data,
    test_size=0.2,
    random_state=42,
    stratify=train_val_data['NAFLD_Status']
)

print(f"\n‚úì Second split (from 4090):")
print(f"  ‚Ä¢ Training: {len(train_data)} records")
print(f"  ‚Ä¢ Validation: {len(val_data)} records")

# Check class distribution in splits
print(f"\nüìä Class Distribution Check:")
print(f"  Training - Absent: {(train_data['NAFLD_Status']==0).sum()}, Present: {(train_data['NAFLD_Status']==1).sum()}")
print(f"  Validation - Absent: {(val_data['NAFLD_Status']==0).sum()}, Present: {(val_data['NAFLD_Status']==1).sum()}")
print(f"  Final Test - Absent: {(final_test_data['NAFLD_Status']==0).sum()}, Present: {(final_test_data['NAFLD_Status']==1).sum()}")

# Save datasets
train_data.to_csv('data/train_data.csv', index=False)
val_data.to_csv('data/val_data.csv', index=False)
final_test_data.to_csv('data/final_test_1000_RESERVED.csv', index=False)
print(f"\n‚úì Datasets saved to 'data/' directory")
print(f"  ‚ö†Ô∏è  'final_test_1000_RESERVED.csv' will be used ONLY after Stage 2")

# ============================================================================
# STEP 5: FEATURE PREPARATION
# ============================================================================
print("\n" + "="*80)
print(" STEP 5: FEATURE PREPARATION ".center(80, "="))
print("="*80)

# Selected features (7 features)
features = ['Gender', 'Age', 'BMI', 'ALT', 'GGT', 'Triglycerides', 'Uric_Acid']
target = 'NAFLD_Status'

print(f"\nüéØ Selected Features (7 features):")
for i, feature in enumerate(features, 1):
    print(f"  {i}. {feature}")

# Prepare datasets (ONLY train and validation - NO final test)
X_train = train_data[features]
y_train = train_data[target]

X_val = val_data[features]
y_val = val_data[target]

print(f"\nüìä Data Shapes:")
print(f"  ‚Ä¢ X_train: {X_train.shape}")
print(f"  ‚Ä¢ X_val: {X_val.shape}")
print(f"  ‚Ä¢ Final test data: RESERVED (not loaded)")

# ============================================================================
# STEP 6: FEATURE SCALING
# ============================================================================
print("\n" + "="*80)
print(" STEP 6: FEATURE SCALING ".center(80, "="))
print("="*80)

print("\n‚öôÔ∏è Applying StandardScaler (important for KNN)...")

# Initialize and fit scaler on training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

print(f"‚úì Features scaled")
print(f"  ‚Ä¢ Mean: ~0, Std: ~1")

# Save scaler for future use
joblib.dump(scaler, 'saved_models/scaler_level1.pkl')
print(f"‚úì Scaler saved: 'saved_models/scaler_level1.pkl'")

# ============================================================================
# STEP 7: MODEL TRAINING
# ============================================================================
print("\n" + "="*80)
print(" STEP 7: TRAINING MODELS ".center(80, "="))
print("="*80)

import time

def train_model_with_timer(model, X_train, y_train, model_name):
    """Train model and track time"""
    print(f"\nüîÑ Training {model_name}...")
    start_time = time.time()
    model.fit(X_train, y_train)
    elapsed = time.time() - start_time
    print(f"   ‚úì Completed in {elapsed:.2f} seconds")
    return model

# Initialize models
print("\nü§ñ Initializing models...")

# 1. K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5, weights='uniform', metric='euclidean')

# 2. Random Forest
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=20,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

# 3. Decision Tree
dt_model = DecisionTreeClassifier(
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    criterion='gini',
    random_state=42
)

# Train all models
knn_model = train_model_with_timer(knn_model, X_train_scaled, y_train, "KNN")
rf_model = train_model_with_timer(rf_model, X_train_scaled, y_train, "Random Forest")
dt_model = train_model_with_timer(dt_model, X_train_scaled, y_train, "Decision Tree")

print(f"\n‚úì All models trained successfully!")

# Save models
joblib.dump(knn_model, 'saved_models/level1_knn.pkl')
joblib.dump(rf_model, 'saved_models/level1_rf.pkl')
joblib.dump(dt_model, 'saved_models/level1_dt.pkl')
print(f"‚úì Models saved in 'saved_models/' directory")

# ============================================================================
# STEP 8: MODEL EVALUATION ON VALIDATION SET ONLY
# ============================================================================
print("\n" + "="*80)
print(" STEP 8: MODEL EVALUATION (Validation Set ONLY) ".center(80, "="))
print("="*80)

print("\n‚ö†Ô∏è  IMPORTANT: Evaluating on VALIDATION SET only")
print("   Final test (1000 records) will be used after Stage 2 is complete\n")

def evaluate_model_detailed(model, X, y, model_name, dataset_name="Validation"):
    """Comprehensive evaluation function"""
    
    # Predictions
    y_pred = model.predict(X)
    y_pred_proba = model.predict_proba(X)[:, 1] if hasattr(model, 'predict_proba') else None
    
    # Metrics
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred, average='binary', zero_division=0)
    recall = recall_score(y, y_pred, average='binary', zero_division=0)
    f1 = f1_score(y, y_pred, average='binary', zero_division=0)
    
    print(f"\n{'='*60}")
    print(f" {model_name} - {dataset_name} Set ".center(60))
    print(f"{'='*60}")
    print(f"  Accuracy:  {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1-Score:  {f1:.4f}")
    
    # ROC-AUC if probability available
    if y_pred_proba is not None:
        roc_auc = roc_auc_score(y, y_pred_proba)
        print(f"  ROC-AUC:   {roc_auc:.4f}")
    
    # Classification Report
    print(f"\n  Classification Report:")
    print(classification_report(y, y_pred, target_names=['Absent (0)', 'Present (1)'], digits=4))
    
    # Confusion Matrix
    cm = confusion_matrix(y, y_pred)
    
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Absent', 'Present'],
                yticklabels=['Absent', 'Present'])
    plt.title(f'{model_name} - Confusion Matrix\n{dataset_name} Set')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.tight_layout()
    plt.savefig(f'plots/level1_confusion_matrices/{model_name.lower().replace(" ", "_")}_{dataset_name.lower()}.png', dpi=300)
    plt.show()
    
    return {
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'ROC-AUC': roc_auc if y_pred_proba is not None else None
    }

# Evaluate all models on validation set ONLY
results_val = []
results_val.append(evaluate_model_detailed(knn_model, X_val_scaled, y_val, "KNN", "Validation"))
results_val.append(evaluate_model_detailed(rf_model, X_val_scaled, y_val, "Random Forest", "Validation"))
results_val.append(evaluate_model_detailed(dt_model, X_val_scaled, y_val, "Decision Tree", "Validation"))

# ============================================================================
# STEP 9: MODEL COMPARISON
# ============================================================================
print("\n" + "="*80)
print(" STEP 9: MODEL COMPARISON ".center(80, "="))
print("="*80)

# Create comparison DataFrame
comparison_df = pd.DataFrame(results_val)
comparison_df = comparison_df.set_index('Model')

print(f"\nüìä Model Performance Comparison (Validation Set):")
print(comparison_df.round(4))

# Find best model
best_model_name = comparison_df['Accuracy'].idxmax()
best_accuracy = comparison_df['Accuracy'].max()

print(f"\nüèÜ Best Model: {best_model_name}")
print(f"   Accuracy: {best_accuracy:.4f} ({best_accuracy*100:.2f}%)")

# Save comparison
comparison_df.to_csv('results/level1_model_comparison_validation.csv')
print(f"\n‚úì Comparison saved: 'results/level1_model_comparison_validation.csv'")

# Visualize comparison
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

for idx, metric in enumerate(metrics):
    axes[idx].bar(comparison_df.index, comparison_df[metric], color=['#3498db', '#2ecc71', '#e74c3c'])
    axes[idx].set_title(f'{metric}')
    axes[idx].set_ylim([0.90, 1.03])
    axes[idx].set_ylabel('Score')
    axes[idx].tick_params(axis='x', rotation=45)
    
    # Add value labels on bars
    for i, v in enumerate(comparison_df[metric]):
        axes[idx].text(i, v + 0.02, f'{v:.3f}', ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('plots/level1_model_comparison_validation.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# STEP 10: FEATURE IMPORTANCE (for RF and DT)
# ============================================================================
print("\n" + "="*80)
print(" STEP 10: FEATURE IMPORTANCE ".center(80, "="))
print("="*80)

def plot_feature_importance(model, feature_names, model_name):
    """Plot feature importance for tree-based models"""
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]
        
        print(f"\nüìä Feature Importance - {model_name}:")
        for i, idx in enumerate(indices, 1):
            print(f"  {i}. {feature_names[idx]}: {importances[idx]:.4f}")
        
        plt.figure(figsize=(10, 6))
        plt.bar(range(len(importances)), importances[indices], color='steelblue')
        plt.xticks(range(len(importances)), [feature_names[i] for i in indices], rotation=45, ha='right')
        plt.xlabel('Features')
        plt.ylabel('Importance')
        plt.title(f'Feature Importance - {model_name}')
        plt.tight_layout()
        plt.savefig(f'plots/level1_feature_importance/{model_name.lower().replace(" ", "_")}_importance.png', dpi=300)
        plt.show()

# Plot for Random Forest and Decision Tree
plot_feature_importance(rf_model, features, "Random Forest")
plot_feature_importance(dt_model, features, "Decision Tree")

# ============================================================================
# STEP 11: ROC CURVES (Validation Set)
# ============================================================================
print("\n" + "="*80)
print(" STEP 11: ROC CURVES (Validation Set) ".center(80, "="))
print("="*80)

plt.figure(figsize=(10, 8))

models = {
    'KNN': knn_model,
    'Random Forest': rf_model,
    'Decision Tree': dt_model
}

colors = ['#3498db', '#2ecc71', '#e74c3c']

for (model_name, model), color in zip(models.items(), colors):
    y_pred_proba = model.predict_proba(X_val_scaled)[:, 1]
    fpr, tpr, _ = roc_curve(y_val, y_pred_proba)
    roc_auc = roc_auc_score(y_val, y_pred_proba)
    
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.3f})', 
             color=color, linewidth=2)

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier', linewidth=1)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curves - All Models (Testing Set)', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=11)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('plots/level1_roc_curves/roc_curves_validation.png', dpi=300)
plt.show()

print(f"‚úì ROC curves saved")

# ============================================================================
# STEP 12: SAVE BEST MODEL INFO
# ============================================================================
print("\n" + "="*80)
print(" STEP 12: SAVE BEST MODEL INFO ".center(80, "="))
print("="*80)

# Save best model name
best_model_info = {
    'best_model_name': best_model_name,
    'best_accuracy': best_accuracy,
    'all_results': comparison_df.to_dict()
}

import json
with open('results/level1_best_model_info.json', 'w') as f:
    json.dump(best_model_info, f, indent=4)

print(f"\n‚úì Best model info saved: 'results/level1_best_model_info.json'")
print(f"  Best Model: {best_model_name}")
print(f"  Validation Accuracy: {best_accuracy:.4f}")

# ============================================================================
# FINAL SUMMARY
# ============================================================================
print("\n" + "="*80)
print(" STAGE 1 COMPLETE ".center(80, "="))
print("="*80)

print(f"\n‚úÖ Stage 1 Training Complete!")
print(f"\nüìä Summary:")
print(f"  ‚Ä¢ Training records used: {len(train_data)}")
print(f"  ‚Ä¢ Validation records used: {len(val_data)}")
print(f"  ‚Ä¢ Models trained: KNN, Random Forest, Decision Tree")
print(f"  ‚Ä¢ Best model: {best_model_name} ({best_accuracy*100:.2f}%)")
print(f"\n‚ö†Ô∏è  RESERVED for Later:")
print(f"  ‚Ä¢ Final test set: 1000 records (NOT USED YET)")
print(f"  ‚Ä¢ Will be tested AFTER Stage 2 is complete")
print(f"\nüìÅ Saved Files:")
print(f"  ‚Ä¢ Models: saved_models/level1_*.pkl")
print(f"  ‚Ä¢ Scaler: saved_models/scaler_level1.pkl")
print(f"  ‚Ä¢ Results: results/level1_model_comparison_validation.csv")
print(f"  ‚Ä¢ Plots: plots/level1_*/")
print(f"  ‚Ä¢ Reserved test data: data/final_test_1000_RESERVED.csv")
print(f"\n‚è≠Ô∏è  Next Steps:")
print(f"  1. Build Stage 2 (Severity Classification)")
print(f"  2. Integrate Stage 1 + Stage 2 into hierarchical system")
print(f"  3. Test complete system on 1000 reserved records")
print(f"  4. Add explainability (SHAP)")

print("\n" + "="*80)
print(" END OF STAGE 1 ".center(80, "="))
print("="*80)
```
